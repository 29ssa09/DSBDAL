```python
# Name:Gajanan Narwade
import requests
from bs4 import BeautifulSoup as bs
```

```python
link = 'https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
page = requests.get(link)
```

```python
page
```

```python
page.content
```

```python
# parse the page content in html form
soup = bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
#revier name
names = soup.find_all('span',class_='a-profile-name')
```

```python
names
```

```python
cust_name = []
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
cust_name
```

```python
cust_name.pop(0) # removing duplicate entry
```

```python
#Finding titles of all reviews
title = soup.find_all('a',class_='review-title-content')
```

```python
title
```

```python
review_title = []
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
# stripping off newline at the beginning
review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title
```

```python
# stripping off newline at the end
review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

```python
#Finding all ratings
rating = soup.find_all('i',class_='review-rating')
rating
```

```python
rate = []
for i in range(0,len(rating)):
    rate.append(rating[i].get_text())
rate
```

```python
#finding review body
review = soup.find_all("span",{"data-hook":"review-body"})
review
```

```python
review_content = []
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content
```

```python
#Stripping newline character at the beginning and the end
review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
review_content
```

```python
review_content[:] = [reviews.rstrip('\n') for reviews in review_content]
review_content
```

```python
# adding remaining columns
if len(review_title) < len(df.index):
    review_title.extend([None] * (len(df.index) - len(review_title)))

if len(rate) < len(df.index):
    rate.extend([None] * (len(df.index) - len(rate)))

if len(review_content) < len(df.index):
    review_content.extend([None] * (len(df.index) - len(review_content)))

df['Review title']=review_title
df['Ratings']=rate
df['Reviews']=review_content
```
